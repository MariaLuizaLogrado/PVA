```python
from dtaidistance import dtw
import warnings
warnings.filterwarnings("ignore")
resultados = []

for i in range(len(df_pixels)):
    for j in range(i+1, len(df_pixels)):
        dist = dtw.distance(df_pixels.iloc[i][1], df_pixels.iloc[j][1])
        resultados.append({"tag": df_pixels.iloc[i][0], "compare":df_pixels.iloc[j][0], "dist":dist})
        resultados.append({"tag": df_pixels.iloc[j][0], "compare":df_pixels.iloc[i][0], "dist":dist})
        
df_resultados = pd.DataFrame(resultados)
df_resultados.to_csv('dtw_distances.csv')
```

```python
from dtaidistance import dtw
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'teste': [np.random.randint(0, 100, 5) for _ in range(10)]
})

resultados = []

for i in range(len(df)):
    for j in range(i+1, len(df)):
        dist = dtw.distance(df.iloc[i]['teste'], df.iloc[j]['teste'])
        resultados.append({
            'linha_1': i,
            'linha_2': j,
            'dtw_distance': dist
        })
        resultados.append({
            'linha_1': j,
            'linha_2': i,
            'dtw_distance': dist
        })
        print(f"Distância DTW entre linha {i} e linha {j}: {dist}")

df_resultados = pd.DataFrame(resultados)
df_resultados.to_csv('dtw_distances.csv', index=False)
```
